```python
import numpy as np
import time
import matplotlib.pyplot as plt

sizes = np.logspace(2, 8, num=7, dtype=int)
mflops = []
mem_kb = []
reps = 100

for n in sizes:
    mat = np.random.rand(1, n)
    start = time.perf_counter()
    for _ in range(reps):
        out = 2.0 * mat
    end = time.perf_counter()
    ops = n * reps
    perf = (ops / (end - start)) / 1e6
    mflops.append(perf)
    mem_kb.append((n * 8) / 1024)

plt.loglog(mem_kb, mflops, 'o-')
plt.xlabel("Vector size [KB]")
plt.ylabel("Performance [MFLOP/s]")
plt.title("Row Doubling Performance")
plt.tight_layout()
plt.savefig("row_scaling.png")
plt.show()
```
